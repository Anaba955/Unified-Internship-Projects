```python
df['Sales to Mar Ratio'] = df['Sales Qtr - Crore'] / df['Mar Cap - Crore']
df
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Mar Cap - Crore'], df['Sales Qtr - Crore'], color='blue')
plt.ylabel('Sales')
plt.xlabel('Market Cap (in billions)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Basic statistics
market_cap_mean = df['Mar Cap - Crore'].mean()
sales_qtr_mean = df['Sales Qtr - Crore'].mean()

print(f"Mean Market Cap: {market_cap_mean}")
print(f"Mean Sales Qtr: {sales_qtr_mean}")

# Distribution of Market Cap
plt.figure(figsize=(10, 6))
sns.histplot(df['Mar Cap - Crore'], bins=30, kde=True)
plt.title('Distribution of Market Cap')
plt.xlabel('Market Cap (Crore)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Distribution of Sales Qtr
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales Qtr - Crore'], bins=30, kde=True)
plt.title('Distribution of Sales Qtr')
plt.xlabel('Sales Qtr (Crore)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Market Cap vs. Sales Qtr
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Mar Cap - Crore', y='Sales Qtr - Crore', data=df)
plt.title('Market Cap vs. Sales Qtr')
plt.xlabel('Market Cap (Crore)')
plt.ylabel('Sales Qtr (Crore)')
plt.grid(True)
plt.show()

# Top 10 Companies by Market Cap
top_10_market_cap = df.nlargest(10, 'Mar Cap - Crore')

plt.figure(figsize=(12, 8))
sns.barplot(x='Mar Cap - Crore', y='Name', data=top_10_market_cap)
plt.title('Top 10 Companies by Market Cap')
plt.xlabel('Market Cap (Crore)')
plt.ylabel('Company')
plt.grid(True)
plt.show()

# Top 10 Companies by Sales Qtr
top_10_sales_qtr = df.nlargest(10, 'Sales Qtr - Crore')

plt.figure(figsize=(12, 8))
sns.barplot(x='Sales Qtr - Crore', y='Name', data=top_10_sales_qtr)
plt.title('Top 10 Companies by Sales Qtr')
plt.xlabel('Sales Qtr (Crore)')
plt.ylabel('Company')
plt.grid(True)
plt.show()
```
